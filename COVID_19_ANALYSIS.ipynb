```python
country=world.groupby(['Country/Region'],as_index=False).value_counts()
country
```

```python
sns.set_style("darkgrid")
sns.barplot(y=country['Population'],x=country['ActiveCases'],hue=country['Continent'])
plt.show()
```

```python
sns.set_style("dark")
fig,ax=plt.subplots(1,2,figsize=(10,10))
ax[0 ].plot(country['TotalCases'],country['TotalDeaths'])
ax[0].set_title("Total Cases vs Total Deaths")
ax[0].set_xlabel("Total Cases")
ax[0].set_ylabel("Total Deaths")

ax[1].pie(country['NewDeaths'])
ax[1].legend(country['NewDeaths'])
ax[1].set_title("New Deaths")


```
